# Data Visualization

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=5, cols=1)

fig.add_trace(go.Scatter( x= data_subset["date"], y=data_subset["grocery_and_pharmacy_percent_change_from_baseline"], 
                         name = 'grocery_and_pharmacy_percent_change_from_baseline'),
              row = 1, col = 1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["parks_percent_change_from_baseline"], 
               name = 'parks_percent_change_from_baseline'),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["transit_stations_percent_change_from_baseline"], 
               name = 'transit_stations_percent_change_from_baseline'),
    row=3, col=1
)


fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["workplaces_percent_change_from_baseline"], 
               name = 'workplaces_percent_change_from_baseline' ),
    row=4, col=1
)

fig.add_trace(
    go.Scatter(x=data_subset["date"], y=data_subset["residential_percent_change_from_baseline"], 
              name = 'residential_percent_change_from_baseline'),
    row=5, col=1
)

fig.update_layout(height=1800, width=1200)
fig.show()
```
